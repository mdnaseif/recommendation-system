## Applying Basic algorithm

```python
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def standardize(row):
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row

df_std = data.apply(standardize).T
df_std
```

```python
df_std.describe()
```

```python
sparse_df = sparse.csr_matrix(df_std.values)
corrMatrix = pd.DataFrame(cosine_similarity(sparse_df),index=data.columns,columns=data.columns)
corrMatrix
```

```python
corrMatrix = data.corr(method='pearson')
corrMatrix
```

## Try the system with 1 and multi items

```python
mean = 1.5
#make prediction
def get_similar(item_name,rating):
    similar_score = corrMatrix[item_name]*(rating-1.5)
    similar_score = similar_score.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_score
```

```python
get_similar("Qadir",1)
```

```python
action_lover = [("Cold_Brew",2),("Ikleer",2),("Costarica",2)]
similar_scores = pd.DataFrame()
for item,rating in action_lover:
    similar_scores = similar_scores.append(get_similar(item,rating),ignore_index = True)

similar_scores.head(10)
```
